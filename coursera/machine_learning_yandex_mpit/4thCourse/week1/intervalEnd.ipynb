```python
np.random.seed(0)

def bootstrap_sample_generate(sample, n_samples):
    indices = np.random.randint(0, len(sample), (n_samples, len(sample)))
    data = np.array(sample)[indices]
    
    return data

def stat_intervals(array, alpha):
    boundaries = np.percentile(array, [100 * alpha / 2, 100 * (1 - alpha / 2)])
    return boundaries
```
