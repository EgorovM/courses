```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

## Равномерное распределение

### Границы распределения: 0..100

```python
a = 0
b = 100

# обратите внимание, что в этой функции задается левая граница и масштаб, а не левая и правая границы:
uniform_rv = sts.uniform(a, b-a)

sample = uniform_rv.rvs(1000)
```

```python
# Гистограмма выборки:
plt.hist(sample, normed=True) # Построение графика

# на том же графике построим теоретическую плотность распределения:
x = np.linspace(0, 100, 1000) # Массив значений аргумента
pdf = uniform_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5) 
plt.legend()
plt.ylabel('$fraction of samples\ /f(x)$') 
plt.xlabel('$x$') 
plt.show() 

```

## Выборки объёмом 5, 10, 50
### Их гистограммы и соответствующие плотности нормального распрделения

```python
def draw_result(size):
    print 'n=%d:' % size
    # Гистограмма выборки:
    sample = [np.mean(uniform_rv.rvs(size=size)) for _ in range(1000)]
    plt.hist(sample,label='histogram', normed=True) # Построение графика
    
    # на том же графике построим теоретическую плотность распределения:
    mu = (a + b) / 2.0
    sigma = np.sqrt(((b-a)**2 / 12.0) / size)
    
    norm_rv = sts.norm(loc=mu, scale=sigma)
    x = np.linspace(a, b, 1000) # Массив значений аргумента
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, label='theoretical pdf', alpha=0.5) # Построение графика
    plt.legend()
    plt.ylabel('$fraction of samples\ /\ f(x)$') # Метка по оси y в формате TeX
    plt.xlabel('$x$') # Метка по оси x в формате TeX
    plt.show() # Показать график
```

```python
grid_sizes = [5, 10, 50]
for i in grid_sizes:
    draw_result(i)
```

-> Распределение выборочных средних достаточно хорошо описывается нормальным распределением, что является утверждением центральной предельной теоремы.

-> С ростом n точность нормальной аппроксимации увеличивается.

## Потверждение для случая n = 100

```python
i = 100
draw_result(i)
```
