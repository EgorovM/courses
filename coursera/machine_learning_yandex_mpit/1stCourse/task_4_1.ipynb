```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Непрерывное распределение $\chi^2$

Возьмем распределение $\chi^2$ с количеством степеней свободы равным 3 и сгенерируем выборку в 1000 значений

```python
sample = sts.chi2.rvs(3, size=1000)
```

Гистограмма выборки вместе с графиком теоретической функцией плотности распределения

```python
x = np.linspace(0,20,100)
pdf = sts.chi2.pdf(x, 3)
plt.plot(x, pdf, label='theoretical PDF')
plt.hist(sample, bins = 30, density=True, label='sample histogram')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

## Распределение среднего выборки размера 5

Сгенерируем 1000 выборок в $n=5$ значений из взятого распределения $\chi^2$ и вычислим среднее значение для каждой выборки

```python
sample_m = []
for i in range(1000):
    sample_m.append(np.mean(sts.chi2.rvs(3, size=5)))
```

Поскольку для исходного распределения $\chi^2$ матожидание и дисперсия выражаются по формулам $M=k$ и $D=2k$, где k - число степеней свободы, то согласно ЦПТ средние значения выборок будут распределены нормально со значениями $\mu=k$ и $\sigma^2=2k / n$

Построим гистограмму средних значений и теоретическую функцию плотности распределения $N(k, 2k/n)$ с учетом того, что число степеней свободы взятого распределения $\chi^2$ равно 3, а $n = 5$

```python
x = np.linspace(0,10,100)
pdf = sts.norm(3.,np.sqrt(2.* 3./5)).pdf(x)
plt.plot(x, pdf, label='theoretical PDF')
plt.hist(sample_m, bins = 20, density=True, label='sample mean histogram')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```
